```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
crimes_committed_IPC=pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012.csv')
```

```python
crimes_committed_IPC.head()
```

```python
Population_data=pd.read_csv('Population_data.csv')   #
```

```python
Population_data.head()
```

```python
crimes_committed_IPC_M=pd.merge(Population_data,crimes_committed_IPC, on='STATE/UT')
```

```python
crimes_committed_IPC_M.head()
```

```python
crimes_committed_IPC_M.info()
```

```python
crimes_committed_IPC_M.isnull().sum()
```

```python
crimes_committed_IPC_M.shape
```

# 2.1 Analysis of Literacy Rate vs Total Crimes Committed- IPC

```python
plt.figure(figsize=(10,4))
plt.bar(crimes_committed_IPC_M['STATE/UT'],crimes_committed_IPC_M['LITERACY'],color='purple')
plt.xlabel('STATE/UT')
plt.ylabel('LITERACY')
plt.xticks (rotation=90)
```

```python
plt.figure(figsize=(10,4))
plt.bar(crimes_committed_IPC_M['LITERACY'],crimes_committed_IPC_M['TOTAL IPC CRIMES'])
plt.xlabel('LITERACY')
plt.ylabel('TOTAL IPC CRIMES')
```

```python
IPC_CRIME= pd.DataFrame(crimes_committed_IPC_M.groupby(['LITERACY'])['TOTAL IPC CRIMES'].sum().reset_index())
IPC_CRIME
```

```python
sns.scatterplot(x='LITERACY',y='TOTAL IPC CRIMES',data=crimes_committed_IPC_M,color="red")
```

# 2.2 Analysis of the type of crime vs each state vs Literacy rate -IPC

```python
CEL= pd.DataFrame(crimes_committed_IPC_M.groupby(['STATE/UT'])['MURDER','ATTEMPT TO MURDER','CULPABLE HOMICIDE NOT AMOUNTING TO MURDER','RAPE','CUSTODIAL RAPE','OTHER RAPE','KIDNAPPING & ABDUCTION','KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS','KIDNAPPING AND ABDUCTION OF OTHERS','DACOITY'
,'PREPARATION AND ASSEMBLY FOR DACOITY','ROBBERY','BURGLARY','THEFT','AUTO THEFT','OTHER THEFT','RIOTS','CRIMINAL BREACH OF TRUST','CHEATING','COUNTERFIETING','ARSON','HURT/GREVIOUS HURT','DOWRY DEATHS','ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY','INSULT TO MODESTY OF WOMEN','CRUELTY BY HUSBAND OR HIS RELATIVES','IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES','CAUSING DEATH BY NEGLIGENCE','LITERACY' ].sum().reset_index())
CEL
```

# 2.3 Analysis of year-on-year total crime rate.

```python
YWC= pd.DataFrame(crimes_committed_IPC_M.groupby(['YEAR'])['TOTAL IPC CRIMES'].sum().reset_index())
YWC
```

```python
plt.figure(figsize=(10,4))
plt.bar(crimes_committed_IPC_M['YEAR'],crimes_committed_IPC_M['TOTAL IPC CRIMES'])
plt.xlabel('YEAR')
plt.ylabel('TOTAL IPC CRIMES')
```

# 2.4 Analysis of area vs overall crime- IPC

```python
AO= pd.DataFrame(crimes_committed_IPC_M.groupby(['STATE/UT','AREA(KM2)'])['TOTAL IPC CRIMES'].sum().reset_index())
AO
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(y='STATE/UT', x='AREA(KM2)', hue='TOTAL IPC CRIMES', data=AO)
plt.title('STATE/UT VS AREA(KM2)',fontsize=18)
plt.show()
```

# 2.5 Analysis of Population vs overall Crime- IPC

```python
PO= pd.DataFrame(crimes_committed_IPC_M.groupby(['POPULATION'])['TOTAL IPC CRIMES'].sum().reset_index())
PO
```

```python
SPO= pd.DataFrame(crimes_committed_IPC_M.groupby(['STATE/UT','POPULATION'])['TOTAL IPC CRIMES'].sum().reset_index())
SPO
```
