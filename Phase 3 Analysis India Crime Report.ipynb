```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
rapes=pd.read_csv('rapes.csv')
rapes
```

# Store the above data (the result of 3.2) in DataFrame and analyze districts that appear 3 or more than 3 years and print the corresponding state/UT, district, murders, and year in descending order.

```python
rp= pd.DataFrame(rapes.groupby(['STATE_UT','district', 'year'])['rape'].sum().sort_values(ascending=False))
rp
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='district',y='rape',data=rp,color='purple')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='STATE_UT',y='rape',data=rp,color="red")
plt.xticks(rotation=90)
```

```python
kidnapping=pd.read_csv('Kidnapping_and_Abduction.csv')
kidnapping
```

```python
K=kidnapping['Kidnapping_and_Abduction'].value_counts()
K
```

```python
K.shape
```

```python
kdnp= pd.DataFrame(kidnapping.groupby(['STATE_UT','district', 'year'])['Kidnapping_and_Abduction'].sum().sort_values(ascending=False))
kdnp
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='district',y='Kidnapping_and_Abduction',data=kdnp,color="red")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='STATE_UT',y='Kidnapping_and_Abduction',data=kdnp,color="red")
plt.xticks(rotation=75)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='STATE_UT',y='Kidnapping_and_Abduction',data=kdnp,color="red")
plt.xticks(rotation=75)
```
